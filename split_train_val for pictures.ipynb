```python
# Transform: filename--labels  -->   filename--train/val--labels with 0.7:0.3 random split
# python 3.8.3

# import necessary libraries

import os
import random
import shutil
from shutil import copy2

# your labels:
category = ['M.Beer', 'MD.Diet', 'MD.Orig', 'P.Cherry', 'P.diet', 'P.Orig', 'P.Rsugar', 'P.Zero']

# your desired outcome path
saveTrainPATH = r'D:\DM996 MLDL\DM996_S2.2\DM996_S2.2\Soda_Bottles\traintest\train'
saveValPATH = r'D:\DM996 MLDL\DM996_S2.2\DM996_S2.2\Soda_Bottles\traintest\val'

# main part:
if __name__=="__main__":
    
    for i in range(8):
        PATH = r'D:\DM996 MLDL\DM996_S2.2\DM996_S2.2\Soda_Bottles\Soda Bottles' + '/' + category[i]  # go through all labels paths
        
        datafiles = os.listdir(PATH)
        num_image = len(datafiles)        # get the image number under a certain label 
        index_list = list(range(num_image))   # index those images waiting for shuffle
        random.shuffle(index_list)         #random shuffle index (this is where the random component come from for your outcome)
        num = 0
            
        trainDir = saveTrainPATH + '/' + category[i]    # train direction: if target direction don't exist, build a new one.
        if not os.path.exists(trainDir):
            os.mkdir(str(trainDir))
                 
        valDir = saveValPATH + '/' + category[i]      # val direction: if target direction don't exist, build a new one.
        if not os.path.exists(valDir):
            os.mkdir(str(valDir))
        
        # go through index_list copy 70% dataset into train and 30% into val.
        for j in index_list:
            fileName = os.path.join(PATH, datafiles[j])
            if num < num_image*0.7:
                copy2(fileName, trainDir)
            else:
                copy2(fileName, valDir)
            num += 1  

# check outcome
# train size
trainsize = 0
for i in range(8):
    PATHtrain = r'D:\DM996 MLDL\DM996_S2.2\DM996_S2.2\Soda_Bottles\traintest\train' + '/' + category[i]
    trainsize += len(os.listdir(PATHtrain))
print(trainsize)

#val size
valsize = 0
for i in range(8):
    PATHval = r'D:\DM996 MLDL\DM996_S2.2\DM996_S2.2\Soda_Bottles\traintest\test' + '/' + category[i]
    valsize += len(os.listdir(PATHval))
print(valsize)
```
